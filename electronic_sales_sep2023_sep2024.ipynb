# 가설1 : 컴퓨터를 가장많이 구매하는 나이대는 50대, 스마트폰을 많이 구매하는 나이대는 20대 일 것이다.

```python
df2.loc[:,'Age_range'] =  df2['Age'].apply(lambda x:(x//10)*10)
for data in df2['Age_range']:
   data
```

나이대 별로 그래프를 그릴 수 있도록 나이대를 구간 범위 컬럼을 추가한다


```python
df2.head(10)
```

```python
df2.info()
```

```python
df2['Product Type'].unique()
```

```python
product_types = ['Smartphone', 'Tablet', 'Laptop', 'Smartwatch', 'Headphones']

grouped_data = df2.groupby(['Age_range', 'Product Type']).size().unstack(fill_value=0)

#나이대 값 [10,20,30,40,50,60,70,80,90]
age_ranges = df2['Age_range'].unique()

#그래프를 그릴 때 사용자의 시선에서 위에서 아래로 보기 편하게 내림차순 정렬
age_ranges = sorted(age_ranges,reverse=True)
print(age_ranges)

#나이대 별로 색상 지정
colors = plt.cm.tab10(np.linspace(0, 1, len(age_ranges)))

# 그래프 크기 설정
fig, ax = plt.subplots(figsize=(10, 6))

# Product Type을 기준으로 X축을 구성하기 위해 막대 위치 설정
bar_width = 0.1  # 막대 너비
index = np.arange(len(product_types))  # X축 위치

# 각 Age_range별로 막대 그래프 그리기
for i, age_range in enumerate(age_ranges):
    counts = grouped_data.loc[age_range]

    ax.barh(index + i * bar_width, counts, bar_width, label=f'Age Range {age_range}', color=colors[i])

ax.set_title('Product Type Counts by Age Range')
ax.set_ylabel('Product Type')
ax.set_xlabel('Count')
ax.set_yticks(index + bar_width * (len(age_ranges) - 1) / 2)
ax.set_yticklabels(product_types, rotation=45)

ax.legend(title='Age Range')

plt.tight_layout()
plt.show()
```

```python
df2['Parsed Accessories'] = df2['Add-ons Purchased'].apply(lambda x: [item.strip() for item in x.split(',')] if pd.notna(x) else [])

accessory_count_by_product = {}

for product_type in df2['Product Type'].unique():
    product_df = df2[df2['Product Type'] == product_type]
    all_accessories = [item for sublist in product_df['Parsed Accessories'] for item in sublist]
    accessory_count = pd.Series(all_accessories).value_counts()
    accessory_count_by_product[product_type] = accessory_count

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for ax, (product_type, accessory_count) in zip(axes.flat, accessory_count_by_product.items()):
    ax.pie(accessory_count, labels=accessory_count.index, autopct='%.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    ax.set_title(f'{product_type}of accessory proportion')

fig.delaxes(axes.flat[5])

plt.tight_layout()
plt.show()
```

#가설5: 평점은 5점대가 많을 것이다

```python
rating_count_by_quantity = df2.groupby('Quantity')['Rating'].size()

# 굴곡 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(rating_count_by_quantity.index, rating_count_by_quantity.values, marker='o', linestyle='-', color='b')

plt.title('Quantity ', fontsize=16)
plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Quantity Count', fontsize=12)
plt.grid(True)
plt.xticks(rating_count_by_quantity.index)  # Quantity의 고유 값들을 x축에 표시
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

전체적인 별점은 3점대가 가장 많고, 2위는 8점, 5점은 5위 정도의 리뷰를 가진다

```python
rating_count = df2.groupby(['Product Type', 'Rating']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))


for product in rating_count.index:
    plt.plot(rating_count.columns, rating_count.loc[product], marker='o', label=product, alpha=0.7)


plt.title('Product of Rating (1~10)', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rating_count.columns)
plt.legend(title='Product Type')
plt.grid(True)
plt.tight_layout()

plt.show()
```

제품군마다 별점을 보니, 랩톱, 타블렛, 스마트 워치는 3점대, 스마트폰은 5점 또는 2점대로 갈리고,  헤드폰은 거의 2점에서 4점대의 평점을 보인다

```python
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
